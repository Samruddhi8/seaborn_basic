```python
import seaborn as sns
```

```python
import numpy as np
```

```python
x=np.array([1,2,3,4,5,6])
y=np.array([2,4,6,8,10,12])
sns.lineplot(x=x,y=y)
```

```python
data=sns.load_dataset("iris")
sns.scatterplot(x="sepal_length",y="sepal_width",data=data)
```

```python
sns.lineplot(x="petal_length",y="petal_width",data=data,hue="species")
```

```python
sns.lineplot(x='sepal_length',y='petal_width',data=data,hue='species',markers='true')
```

```python
data1=sns.load_dataset("titanic")
sns.barplot(x="class",y="fare",data=data1)
```

```python
sns.barplot(x="class",y="fare",data=data1,hue="survived")
```

```python
sns.histplot(x="petal_length",data=data,kde="true")
```

```python
sns.kdeplot(x="petal_width",data=data)
```

```python
sns.kdeplot(data=data, x="sepal_length",hue="species", fill=True, alpha=0.5)
```

```python
sns.pairplot(data=data,hue="species")
```

```python
num=data.select_dtypes(include=[np.number])
corr = num.corr()
sns.heatmap(corr)
```
